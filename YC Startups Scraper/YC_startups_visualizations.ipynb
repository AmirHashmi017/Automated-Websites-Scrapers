## YC Startups Visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

yc_df = pd.read_csv('yc_companies_clean.csv')

yc_df['tags'] = yc_df['tags'].fillna('Unspecified')
yc_df.head()
```

### Top 10 Industry Categories (Most Popular Sectors)

```python
all_tags = yc_df['tags'].str.split(', ').explode()
tag_counts = all_tags.value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=tag_counts.values, y=tag_counts.index, palette='magma')
plt.title('Top 10 Industry Tags in YC Companies', fontsize=14)
plt.xlabel('Number of Companies', fontsize=12)
plt.ylabel('Industry Tag', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

### Top 10 Locations for YC Startups

```python
top_locations = yc_df['location'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_locations.values, y=top_locations.index, palette='rocket')
plt.title('Top 10 Global Locations for YC Startups', fontsize=14)
plt.xlabel('Number of Startups', fontsize=12)
plt.ylabel('Location (City, State/Region)', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

### YC Companies per Batch Year

```python
yc_df['batch_year'] = yc_df['batch'].str.extract(r'(\d{4})').astype(float)

year_counts = yc_df['batch_year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x=year_counts.index, y=year_counts.values, marker='o', color='#ff6600', linewidth=2.5)
plt.fill_between(year_counts.index, year_counts.values, color='#ff6600', alpha=0.1)

plt.title('YC Growth: Number of Companies per Batch Year', fontsize=14)
plt.xlabel('Batch Year', fontsize=12)
plt.ylabel('Number of Companies', fontsize=12)
plt.xticks(year_counts.index.astype(int), rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
